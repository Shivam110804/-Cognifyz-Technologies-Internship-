# LEVEL 1 (Task 1): Top Cuisines
Determine the top three most
common cuisines in the dataset.

Calculate the percentage of
restaurants that serve each of the top
cuisines.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv' , lineterminator = '\n' ) 
```

```python
df.head()
```

```python
cuisine_counts = df['Cuisines'].value_counts()
```

```python
top_cuisines = cuisine_counts.head(3)
```

```python
total_restaurants = len(df)
percentages = (top_cuisines / total_restaurants) * 100
```

```python
import numpy as np

# Define the colors for the bars
colors = ['Red', 'green', 'orange']

# Plot the bar chart with colors
plt.bar(top_cuisines.index, top_cuisines.values, color=colors)

# Add labels and title
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top Three Cuisines')

# Show the chart
plt.show()
```

```python
plt.pie(top_cuisines.values, labels=top_cuisines.index, autopct='%1.1f%%')
plt.title('Top Three Cuisines')
plt.show()
```
